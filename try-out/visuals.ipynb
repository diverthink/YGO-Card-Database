```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
# Datenimport

data = pd.read_csv('data/cards_cleaned.csv', sep="|", keep_default_na=False)
```

```python
data.columns
```

```python
data
```

```python
data_filtered = data.loc[data['attribute'] != 'None']

fig = px.histogram(data_filtered, x='attribute', color='race')

# Achsentitel setzen
fig.update_xaxes(title_text='Attribute')
fig.update_yaxes(title_text='Count')

# Layout anpassen und anzeigen
fig.update_layout(title_text="Histogramm der Attribute")
fig.show()

fig = px.histogram(data, x='frameType', color='race')

# Achsentitel setzen
fig.update_xaxes(title_text='frameType')
fig.update_yaxes(title_text='Count')

# Layout anpassen und anzeigen
fig.update_layout(title_text="Histogramm der Kartentypen")
fig.show()
```
